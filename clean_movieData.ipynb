```python
import pandas as pd
import ast
import json
```

```python
df=pd.read_csv("movies_metadata.csv",sep=',', low_memory=False)
```

```python
df.shape
```

```python
df.head()
```

```python
def convert_to_json(value):
    try:
        if pd.isna(value) or value == '':
            return None
        # ast.literal_eval safely parses Python dict/list strings
        parsed = ast.literal_eval(value)
        return json.dumps(parsed)   # convert to proper JSON string
    except:
        return value  # return as-is if parsing fails
```

```python
json_columns = [
    'belongs_to_collection',
    'genres',
    'production_companies',
    'production_countries',
    'spoken_languages'
]
```

```python
for col in json_columns:
    df[col] = df[col].apply(convert_to_json)
```

```python
string_columns = [
    'original_language', 'original_title', 'title',
    'overview', 'tagline', 'status', 'homepage',
    'imdb_id', 'poster_path', 'release_date'
]
```

```python
for col in string_columns:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].replace('nan', None)
```

```python
df = df.dropna(subset=['id'])           # remove rows with no ID
df = df.drop_duplicates(subset=['id'])  # remove duplicate movies
```

```python
del_col=[
    'belongs_to_collection',
    'homepage',
    'imdb_id',
    'popularity',
    'video',
    'vote_count'
]
```

```python
df=df.drop(columns=del_col)
```
