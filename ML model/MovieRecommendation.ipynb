```python
update_df=df[['title','overview','genres','tagline','production_companies','spoken_languages','vote_average','popularity']]
```

```python
# delete duplicate rows
update_df.drop_duplicates(inplace=True)
```

```python
update_df = update_df.dropna(subset=['title'])
```

```python
update_df['overview']=update_df['overview'].fillna('')
```

```python
update_df['tagline']=update_df['tagline'].fillna('')
```

```python
import ast
```

```python
update_df['genres'] = update_df['genres'].apply(
    lambda x: " ".join([i['name'] for i in ast.literal_eval(x)]) if x != 'NaN' else ''
)
```

```python
update_df['production_companies'] = update_df['production_companies'].apply(
    lambda x: " ".join([i['name'] for i in ast.literal_eval(x)]) if x != 'NaN' else ''
)
```

```python
update_df['spoken_languages'] = update_df['spoken_languages'].apply(
    lambda x: " ".join([i['name'] for i in ast.literal_eval(x)]) if x != 'NaN' else ''
)
```

```python
update_df['tags']=update_df['overview']+update_df['genres']+update_df['tagline']+update_df['production_companies']+update_df['spoken_languages']
```

```python
update_df['tags'][1]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
  text=text.lower()
  text = re.sub(r'[^\w\s]', '', text)
  words=text.split()

  words=[word for word in words if word not in stop_words]
  words=[lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
update_df['tags']=update_df['tags'].apply(preprocess_text)
```

```python
update_df['tags'][1]
```

```python
update_df=update_df.reset_index(drop=True)
```

```python
indices=pd.Series(update_df.index,update_df['title']).drop_duplicates()
```

```python
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(ngram_range=(1,2), max_features=80000,stop_words='english')
```

```python
tfidf_matrix=tfidf.fit_transform(update_df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
  if title not in indices:
    return "Movie not found"
  idx=indices[title]
  sim_scores=cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
  movie_indices=sim_scores.argsort()[::-1][1:n+1]
  return update_df['title'].iloc[movie_indices]
```

```python
recommend('Avatar')
```
